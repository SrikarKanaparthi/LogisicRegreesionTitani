```python
##reading the data from the csv file
df = pd.read_csv("train.csv")
```

```python
##To display the column names
df.columns
```

```python
## I will take the variables sex,Age,pclass,Embarked,Survived,Ticket
X = pd.DataFrame()
X['sex'] = df['Sex']
X['age'] = df['Age']
X['pclass']=df['Pclass']
X['embarked']=df['Embarked']
X['survived'] = df['Survived']
X['ticket']=df['Ticket']
```

```python
## Here in this step im going to drop the missing values
X = X.dropna(axis=0)
```

```python
## i'll take survival as dependent variable
y = X['survived']
X = X.drop(['survived'], axis=1)
```
